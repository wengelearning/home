```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-3,3,100)
y1=2*x+1
plt.plot(x,y1,color='r',linewidth=1.0,linestyle='-')
ax=plt.gca()#get current axis
ax.spines['right'].set_color('none')#去掉右边框
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')#把x轴的刻度设置为bottom
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',0))#设置bottom对应到0点
ax.spines['left'].set_position(('data',0))
x0=2
y0=2*x0+1
plt.scatter(x0,y0,s=50,color='y')#画点,s表示点的大小
plt.plot([x0,x0],[y0,0],'k--',lw=2)#画线
                            # 指向的点       2x+1的位置
plt.annotate(r'$2x+1=%s$'%y0,xy=(x0,y0),xytext=(+50,-50),textcoords='offset points',
             fontsize=16,arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.3'))#箭头的方式，弧度
plt.text(0,6,r'$this\ is\ the\ text$',fontdict={'size':'16','color':'r'})#添加文本的位置，文本内容 字体 颜色
plt.show()
```

```python
x=np.random.normal(0,1,500)#生成符合标准正态分布的一维数据
y=np.random.randn(500)
plt.scatter(x,y,s=30,c='b',alpha=0.5)#alpha表示透明度
plt.show()
```
