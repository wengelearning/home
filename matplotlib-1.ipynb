```python
import matplotlib.pyplot as plt
x=[1,2]
y=[2,4]
plt.scatter(x,y)#散点图
plt.show()#展示绘画框
```

```python
x=[1,2,3,4,5,6,7,8]
y=[2,1,3,5,2,6,12,7]
plt.plot(x,y)#画线
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8]
y=[2,1,3,5,2,6,12,7]
plt.plot(x,y,linestyle='--')#调整线的格式
plt.show()
```

```python
fig=plt.figure()#将绘画框进行对象化
p1=fig.add_subplot(211) #将p1定义为绘画框的子图，
                        #211表示将绘画框划分2行1列，最后的1表示第一幅
x=[1,2,3,4,5,6,7,8]
y=[2,1,3,5,2,6,12,7]
p1.plot(x,y)
p2=fig.add_subplot(212)#第二幅
a=[1,2]
b=[2,4]
p2.scatter(a,b)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D#导入3D包
fig=plt.figure()#将绘画框进行对象化
ax=fig.add_subplot(111,projection='3d')
x=[1,1,2,2]
y=[3,4,4,3]
z=[1,100,1,1]
ax.plot_trisurf(x,y,z)#用函数填满4个点组成的三角形空间
plt.show()
```
