```python
tuples=list(zip(*[['bar','bar','baz','baz','foo','foo','qux','qux'],
                  ['one','two','one','two','one','two','one','two']]))
tuples
```

```python
index=pd.MultiIndex.from_tuples(tuples,names=['first','second'])#多重索引
df=pd.DataFrame(np.random.randn(8,2),index=index,columns=['A','B'])
df2=df[:4]
df2
```

```python
stacks=df2.stack()
stacks
```

```python
stacks.unstack()
```

```python
stacks.unstack(1)
```

```python
stacks.unstack(0)
```

```python
df=pd.DataFrame({'A':['one','one','two','two']*3,
                'B':['A','B','C']*4,
                'C':['foo','foo','foo','bar','bar','bar']*2,
                'D':np.random.randn(12),
                'E':np.random.randn(12)})
df
```

```python
pd.pivot_table(df,values='D',index=['A','B'],columns=['C'])
```

```python
#Times Series
rng=pd.date_range('1/11/2017',periods=100,freq='S')#以S 秒为间隔，(M 月 D 天 H 小时)
ts=pd.Series(np.random.randint(0,500,len(rng)),index=rng)
ts.resample('5S').sum()#按每5秒间隔求和
```

```python
ts.resample('5Min')
```
